```python
import pandas as pd
```

```python
import numpy as np
```

```python
mydata=pd.read_csv(r"C:\Users\lenovo\Desktop\data.csv")
```

```python
X_Feature=mydata[["Quants","Reasoning","English","PI","GD"]]
```

```python
Y_Feature=mydata["Placed?"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_Feature,Y_Feature,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
PlacementPrediction=LinearRegression()
```

```python
PlacementPrediction.fit(X_train,Y_train)
```

```python
value_coefficient=PlacementPrediction.coef_
```

```python
coeffi_dataframe=pd.DataFrame(value_coefficient,X_Feature.columns,columns=["coefficient"])
```

```python
coeffi_dataframe
```

```python
Y_prediction=PlacementPrediction.predict(X_test)
```

```python
Y_prediction
```

```python
result=pd.DataFrame({"Actual":Y_test,"Prediction":Y_prediction})
```
